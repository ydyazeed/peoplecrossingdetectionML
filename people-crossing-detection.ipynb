```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
#Loading the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
model.classes = [0]
img = 'https://previews.123rf.com/images/giuseppemasci/giuseppemasci1605/giuseppemasci160502041/56926133-london-england-january-31-2012-people-crossing-the-street-near-the-abbey-road-studios.jpg'
result = model(img)
%matplotlib inline 
plt.imshow(np.squeeze(result.render()))
plt.show()
```
